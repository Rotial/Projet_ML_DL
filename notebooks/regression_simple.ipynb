# Régression Linéaire simple: Exemple, prédire le prix d'une maison par rapport à sa surface

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("C:/DataProjets/ProjetData1/data/homeprices.csv")
df 
```

```python
plt.xlabel('Area(m^2)')
plt.ylabel('Price(€)')
plt.scatter(df.Area, df.Price, color = 'red',  marker = '+') # méthode de dispersion
```

## Utilisattion du Modèle de Régression Linéaire pour prédire le prix du loyer

```python
reg = LinearRegression() # création d'un objet de regression linéaire
reg.fit(df[['Area']], df.Price) # Ajuster les données, le modèle de prédiction des prix est prêt 
```

```python
#Prédire par exemple le prix d'une maison dont la surface est de 3300m^2
reg.predict([[3300]])
```

## Calcul du coefficient et de l'intercepte: y = ax + b

```python
reg.coef_
```

```python
reg.intercept_
```
