# Regression Lineaire Multiple

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Reading Dataset

```python
dataset = pd.read_csv("C:/DataProjets/ProjetData1/data/50_Startups.csv")
x = dataset.iloc [:,:-1] #La fonction iloc est utilisée pour attribuer les indices à x et y, ':' toutes les lignes, ':-1' sauf la dernière colonne, '-1' considère la dernière colonne
y = dataset.iloc [:, -1].values # 'values' convertir le résultat en un tableau 'numpy'
dataset.head(5)
```

## Transformer les valeurs de la colonne state en numérique

```python
from sklearn.compose import ColumnTransformer # Cette classe permet d'appliquer différentes transformations à différentes colonnes
from sklearn.preprocessing import OneHotEncoder # OneHotEncoder est utilisé pour encoder les variables catégorielles en format binaire
ct = ColumnTransformer( transformers = [('encoder', OneHotEncoder(), [3])], remainder = 'passthrough')# - remainder='passthrough' : garde les autres colonnes inchangées
x = np.array(ct.fit_transform(x))
```

## Diviser les données en ensembles d'entraînement et de test => fonction 'train_test_split'

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2) # test_size = 0.2 signifie que 20% des données seront utilisées pour le test, 80% pour l'entraînement
```

## Implémenter la régression linéaire ordinaire (méthode des moindres carrés)

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

## Prédiction de résultats de l'ensemble de tests 

```python
y_pred = regressor.predict(x_test)
```

```python
df = pd.DataFrame({'Real Values' : y_test, 'Predicted Values' : y_pred})
df
```

## Evaluation de la performance de ce modèle de regressio: EAM, MSE, RMSE

```python
#RMSE
from sklearn import metrics 
print(np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
